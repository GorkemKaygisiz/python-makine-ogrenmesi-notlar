```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv('maaslar.csv')
veriler.head()
```

```python
x = veriler.iloc[:,1:2]
y = veriler.iloc[:,2:3]
#verileri dataframe olarak almak için idexleme yapıyoruz

```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
X= x.values
Y= y.values 
#pandas dataframe'i numpy array'a çevirmek için .values kullanıyoruz


plt.scatter(X,Y,color='blue')
plt.plot(x,lin_reg.predict(X),color='red')
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) #ikinci dereceden polynomial features objesi oluşturduk

x_poly= poly_reg.fit_transform(X)
#X değerini polinomal bi hale çevirdik
x_poly.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color='blue')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=6) #? derecenin artışı grafiğin bükülme kabiliyetini de arttırır.

x_poly= poly_reg.fit_transform(X)
#X değerini polinomal bi hale çevirdik
x_poly.shape

lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color='blue')
```

```python
lin_reg.predict([[11]])
```

```python
lin_reg2.predict(poly_reg.fit_transform([[11]]))
#lin_reg2 eğitilirken 3 sütunlu veri gördüğü için ona her zaman 3 sütunlu veri vermek zorundayız.
#bu yüzden poly_reg ile verilen değeri 3 sütunluk bir veriye çeviriyoruz
```

# non-linear model şablonu

```python
x = veriler.iloc[:,1:2]
y = veriler.iloc[:,2:3]
#verileri dataframe olarak almak için idexleme yapıyoruz

X= x.values
Y= y.values 
#pandas dataframe'i numpy array'a çevirmek için .values kullanıyoruz


from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) #ikinci dereceden polynomial features objesi oluşturduk

x_poly= poly_reg.fit_transform(X)
#X değerini polinomal bi hale çevirdik
x_poly.shape


lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color='blue')
```
