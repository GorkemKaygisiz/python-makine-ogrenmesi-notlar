# POLYNOMIAL REGRESSION
* Polinomal regresyon, veriler arasındaki ilişkinin düz bir çizgiyle değil, bir eğriyle ifade edildiği regresyon türüdür.

$$y = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + \dots + \beta_nx^n + \epsilon$$

* Bu formülde her bir katsayı ilgili terimin sonuç üzerindeki ağırlığını belirler.
* Sondaki E hata payıdır
* Üs sayısı arttıkça grafiğin bükülme yeteneği artar.
