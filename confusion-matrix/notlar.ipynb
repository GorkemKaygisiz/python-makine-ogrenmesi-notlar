# CONFUSION MATRIX
> Bir sınıflandırma modelinin hangi sınıfları doğru, hangilerini yanlış tahmin ettiğini gösteren bir tablodur. 
## BINARY 
* Positive ve Negative terimleri iyi/kötü anlamına gelmiyor.
* Sadece aradığımız şey var mı yok mu sorusunun cevabı.
* Örneğin sağlıklı-hasta örneğinde **Bu kişi hasta mı ?** sorusunu soruyoruz.
    - Positive =`Evet hasta`
    - Negative =`Hayır hasta değil`
* Model doğru söylediğinde **True**
* Model yanlış söylediğinde **False**

## RECALL, F1-SCORE, PRECISION, ACCURACY
* Bu metriklerin hepsi doğrudan confusion matrix'in verdiği sonuçtan ölçülür.
**Precision**
* Tahmin ettiğim pozitifler ne kadar doğru?
**Recall**
* Gerçek pozitifleri ne kadar yakaladım?
**F1-SCORE**
* Precision ve Recall'un dengeli ortalaması
**Accuracy**
* Genel doğruluk
