## Kümeleme (Clustering)
* Clustering, makine öğrenmesinde kullanılan bir 'gözetimsiz öğrenme' (Unsupervised Learning) yöntemidir.
* Temel amacı elde bulunan etiketlenmemiş veride benzer olanları bir araya toplamaktır.
>Kümelemede etiketler önceden bilinmez, algoritma kendi başına gruplar oluşturur.

# Basit bir örnek
Müşteri	Yaş	Aylık Harcama
A	22	1000
B	24	1200
C	25	1100
D	50	7000
E	52	6800
F	48	7200

Küme 1:
A
B
C

Küme 2:
D
E
F

* A,B,C yaş ve harcama açısından benziyor
* D,E,F de kendi aralarında benziyor


# Clustering ve Classification Arasındaki Fark
> Sınıflandırmada veriler etiketlidir, model etiketi öğrenir. Clusteringde ise etiket yoktur.

# Kümeleme Sonucu Nasıl Değerlendirilir ?
- Silhouette Score
- Davies-Bouldin Index
- Calinski-Harabasz Score


# En Popüler Kümeleme Algoritması: K-Means
**Adım 1**
K adet merkez seçilir, örneğin K = 2, 2 adet küme oluşturulacak demektir.

**Adım 2**#
Her veri en yakın merkeze atanır
Küme A ve Küme B olmak üzere

**Adım 3**
Merkezler yeniden hesaplanır. Kümenin ortası bulunur

**Adım 4**
Tekrar atanır, bu işlem sürekli devam eder ve sonunda kümeler sabit hale gelir.


# Hiyerarşik Kümeleme (Hierarchical Clustering)
* Hiyerarşik kümeleme verileri bir ağaç yapısı oluşturacak şekilde gruplandıran bir yöntemdir. K-Means'ten en büyük farkı, başlangıçta '**Kaç küme olsun?**' sorusuna cevap vermek zorunda olmamaktır.
* Süreç bittiğinde elimizde **Dendrogram** adı verilen bir şema oluşur ve bu şemaya bakarak veriyi istediğimiz yerden kesip küme sayısına karar verebiliriz.

## Nasıl Çalışır ?
* Her veri noktası başlangıçta kendi başına bir kümedir. (Örneğin 10 veri varsa 10 küme vardır)
* Birbirine en yakın olan iki küme birleşir ve artık 9 küme kalır.
* Kalan kümeler arasında tekrar en yakın olanlar bulunur ve birleştirilir.
* Bu işlem tüm veriler tek bir büyük kümede toplanana kadar devam eder.

# Basit bir benzetme
* Rastgele evlerin olduğu bir bölge düşünelim. 
* Önce birbirine bitişik iki ev birleşip bir bahçe oluşturur.
* Sonra yakın bahçeler birleşip bir sokak olur.
* Sokaklar birleşip mahalleleri oluşturur.
* Mahalleler ise ilçeyi, ilçe şehri oluşturur.
