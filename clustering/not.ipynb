## Kümeleme (Clustering)
* Clustering, makine öğrenmesinde kullanılan bir 'gözetimsiz öğrenme' (Unsupervised Learning) yöntemidir.
* Temel amacı elde bulunan etiketlenmemiş veride benzer olanları bir araya toplamaktır.
>Kümelemede etiketler önceden bilinmez, algoritma kendi başına gruplar oluşturur.

C	25	1100

E	52	6800

Küme 1:
A
B
C

Küme 2:
D
E
F

* A,B,C yaş ve harcama açısından benziyor
* D,E,F de kendi aralarında benziyor

## K-Means'de Rassal Başlangıç Tuzağı (Random Initialization Trap) Nedir?
- K-Means algoritmasının başlangıç merkezlerini (centroid) rastgele seçmesi nedeniyle, yanlış noktadan başlayıp veriyi hatalı gruplandırması ve en iyi sonucu (`Global Optimum`) kaçırıp kötü ama stabil bir sonuca (`Local Minimum`) takılıp kalmasıdır.
- K-Means 'en yakındakini sahiplen' mantığıyla çalışır, eğer seçilen başlangıç noktaları birbirine çok yakınsa, verinin uzak kısımları sahipsiz kalır veya yanlış kümeye dahil edilir.

## Random Initialization Trap Nasıl Aşılır ? 
**K-Means++**
- Merkezleri seçerken birbirlerinden olabildiğince uzak noktaları seçen akıllı bir başlangıç algoritması kullanmak.

**N-Init**
- Algoritmayı farklı rastgele başlangıçlarla 10-20 kez çalıştırıp içlerinden en düşük hatayı veren sonucu seçmek

## WCSS (Within-Cluster Sum of Squares)
- K-Means kümeleme algoritmasında başarımızı ve kümeleme kalitemizi ölçmek için kullandığımız en temel matematiksel değerdir.
- En basit haliyle küme içindeki noktaların, kendi merkezlerine ne kadar yakın olduğunu ölçer. Eğer küme içindeki noktalar, küme merkezine (centroid) çok yakınsa, o küme sıkı ve başarılı bir kümedir.

**Nasıl Hesaplanır?**
1. Her noktanın bağlı olduğu küme merkezine olan uzaklığı ölçülür.
2. Bu uzaklıkların karesi alınır (negatif değerden kurtulmak ve uzak olan hataları daha çok cezalandırmak için)
3. Tüm noktalar için bu kareler toplanır.
4. WCSS düşük olması istenir.
5. Elbow method ile en ideal K sayısını belirleriz.

# Hiyerarşik Kümeleme (Hierarchical Clustering)
* Hiyerarşik kümeleme verileri bir ağaç yapısı oluşturacak şekilde gruplandıran bir yöntemdir. K-Means'ten en büyük farkı, başlangıçta '**Kaç küme olsun?**' sorusuna cevap vermek zorunda olmamaktır.
* Süreç bittiğinde elimizde **Dendrogram** adı verilen bir şema oluşur ve bu şemaya bakarak veriyi istediğimiz yerden kesip küme sayısına karar verebiliriz.

## Nasıl Çalışır ?
* Her veri noktası başlangıçta kendi başına bir kümedir. (Örneğin 10 veri varsa 10 küme vardır)
* Birbirine en yakın olan iki küme birleşir ve artık 9 küme kalır.
* Kalan kümeler arasında tekrar en yakın olanlar bulunur ve birleştirilir.
* Bu işlem tüm veriler tek bir büyük kümede toplanana kadar devam eder.

# Basit bir benzetme
* Rastgele evlerin olduğu bir bölge düşünelim. 
* Önce birbirine bitişik iki ev birleşip bir bahçe oluşturur.
* Sonra yakın bahçeler birleşip bir sokak olur.
* Sokaklar birleşip mahalleleri oluşturur.
* Mahalleler ise ilçeyi, ilçe şehri oluşturur.
