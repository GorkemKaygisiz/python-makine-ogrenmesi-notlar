# RANDOM FOREST CLASSIFICATION
> Hem sınıflandırma hem de regresyonda kullanılan **ensemble learning** yöntemidir. Temelde birden fazla ağacın bir araya gelerek tek bir güçlü model oluşturması mantığına dayanır.

## RANDOM FOREST NASIL ÇALIŞIR?
* Algoritmanın mantığı `Böl ve Yönet` stratejisine benzeyen şu adımlarla özetlenir;
**1- Veri Örnekleme (Bootstrapping)**
- Algoritma veri setinden rastgele seçimler yaparak birden fazla alt veri kümesi oluşturur. Bu işlem sırasında aynı veri satırı birden fazla kez seçilebilir ve her bir karar ağacı bu farklı alt kümelere göre eğitilir.
**2- Özellik Seçimi (Feature Randomness)**
- Karar ağaçları dallanırken tüm özellikleri kullanmak yerine her düğümde rastgele seçilmiş belirli sayıdaki özellik arasından en iyisini seçer. Bu ağaçların birbirine çok benzemesini engeller ve model çeşitliliğini arttırır.
**3- Bağımsız Ağaçların Oluşturulması**
- Binlerce karar ağacı birbirinden bağımsız olarak eğitilir. Her ağa. kendisine verilen veri setine göre bir tahmin üretir.
**4- Oylama (Majority Voting)**
- En çok hangi sınıf tercih edildiyse, `Random Forest'ın` nihai kararı o olur.


## NEDEN RANDOM FOREST KULLANMALIYIZ?
* Aşırı öğrenmeyi engeller.
* Karmaşık veri setlerinde iyi performans gösterir
* Eksik verilere karşı dayanıklıdır.
