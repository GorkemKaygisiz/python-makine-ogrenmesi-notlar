# DECISION TREE
* Bir sorunu çözmek veya tahmin yürütmek için tıpkı bir ağaç gibi dallara ayrılan bir yapıdır.
* Örnek: Hava yağmurlu mu ? Evet ise yanına şemsiye al, hayır ise şemsiye alma.

## KARAR AĞACI YAPISI
* Root Node (kök düğüm): Aağaçın en tepesindeki ilk sorudur. Verinin bölündüğü ilk noktadır.
* Branches (dallar): Sorulan soruların cevaplarıdır
* Internal Nodes (iç düğümler): Dallar üzerinden gidilen yeni soruların olduğu duraklardır.(Hava güneşliyse, nem oranı yüksek mi?)
* Leaf Nodes (yapraklar): Artık soru sorulmayan, sonucun veya kararın verildiği uç noktalardır.

## DECISION TREE ÖZELLİKLERİ
* Ham veriyle çalışabilir, verileri önceden düzenlemeyek her zaman şart değildir.
* Hem tahmin hem sınıflandırma yapar.
* Karar ağaçlarının en büyük düşmanı overfitting'dir(ezberleme). Eğer ağaç çok fazla dal budak salarsa, eldeki veriyi ezberler ve yeni veri geldiğinde ise hata yapar. Çünkü genel mantığı değil o anki verinin içindeki gürültüleri öğrenmiştir.
* Bunu engellemek için ağaç budanır **(pruning)** veya daha gelişmiş yöntemler olan **Random Forest** gibi birçok ağacın birleştiği yöntemler kullanılır
