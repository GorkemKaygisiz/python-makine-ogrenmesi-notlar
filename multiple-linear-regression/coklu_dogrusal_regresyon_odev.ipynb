```python
import pandas as pd
import numpy as np
veriler =pd.read_csv('play_tennis.csv')
```

```python
veriler.head()
```

```python
#outlook sütunu kategorik olduğu için onu makinenin anlayacağı dile çevireceğiz.

outlookSütunu = veriler.iloc[:,0:1].values
outlookSütunu
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
outlookSütunu[:,0] = le.fit_transform(veriler.iloc[:,0])
#fit -> kaç farklı kategori var öğren.
#transform -> şimdi bunları sayıya çevir.
outlookSütunu

ohe = preprocessing.OneHotEncoder()
outlookSütunları=ohe.fit_transform(outlookSütunu).toarray() 
outlookSütunları
```

```python
outlookSonuc = pd.DataFrame(data=outlookSütunları,index=range(14),columns=['overcast','rainy','sunny'])
outlookSonuc
```

```python
#windy sütunu boolean olduğu için onu da makinenin anlayacağı dile çevireceğiz

windySütunu = veriler.iloc[:,3:4].astype(int).values
le = preprocessing.LabelEncoder()
windySütunu[:,0] = le.fit_transform(veriler.iloc[:,3])
windySütunu


windySonuc = pd.DataFrame(data=windySütunu,index=range(14),columns=['windy'])
windySonuc
```

```python
#play sütunu makineye öğretmek istediğimiz sütun, bunun için sadece labelEncoder yapıcaz.

playSütunu= veriler.iloc[:,4:5].values

le = preprocessing.LabelEncoder()
playSütunu[:,0]= le.fit_transform(veriler.iloc[:,4:5])
playSonuc=playSütunu
playSonuc
```

```python
#datasetteki sayısal sütunları da çekiyoruz
#1 ve 2.indeksler

sayisalSütunlar = veriler.iloc[:,[1,2]]
sayisalSütunlar
```

```python
concatted_df = pd.concat([outlookSonuc,windySonuc,sayisalSütunlar],axis=1)
#axis= 1 yaapmamızın sebebi yan yana birleştirmek istememiz
concatted_df
```

```python
concatted_df = concatted_df.drop('overcast',axis=1)
#axis=1 burada sütun sildiğimizi belirtiyor.
concatted_df
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(concatted_df,playSonuc,test_size=0.33,random_state=0)

#makineye eğitim verilerini verdik, henüz eğitmedik
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(x_train,y_train)
#makineye eğitim verilerini verdim
```

```python
#tahmin etme bölümü

y_pred=regressor.predict(x_test)
y_pred
```
