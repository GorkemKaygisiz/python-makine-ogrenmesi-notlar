# Çoklu Doğrusal Regresyon (Multiple Linear Regression)

* Çoklu Doğrusal Regresyonda tahmin yaparken sonucu etkileyen birden fazla etken vardır. "Çoklu" kavramı da buradan gelir.

* Gerçek hayatta bir evin fiyatını sadece metrekaresi belirlemez. Oda sayısı, bulunduğu kat, şehir merkezine uzaklığı gibi birçok faktör vardır.

## ÖRNEK
Bir öğrencinin sınav notunu tahmin etmek istediğimizi düşünelim. Bakacağımız etkenler şunlar olabilir
* Kaç saat ders çalıştığı, 
* Önceki sınav notlarının ortalaması
* Gece kaç saat uyuduğu
Bu yöntemde her bir etkenin sonuca ne kadar etki edeceği (ağırlığı) hesaplanır. Mesela ders çalışmak notu çok yükseltirken, uykusuz kalmak biraz düşürebilir. Algoritma bu ağırlıkları belirleyerek bize bir formül sunar.

$$Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \beta_3X_3 + ... + \epsilon$$

# KUKLA DEĞİŞKEN (DUMMY VARIABLE)

* Dummy variable (Kukla Değişken), modelin içine dahil edilemeyen metinsel(kategorik) verileri, modelin anlayabileceği sayısal (0 ve 1) formatına dönüştürme yöntemidir.
* Bir modele girdi olarak kadın veya erkek yazamazsınız veya şehirleri matematiksel işlemlerde kullanamazsınız.

## NASIL ÇALIŞIR?
* Dummy variable yöntemi, her bir kategori çin yeni bir sütun oluşturur ve o satır ile ilgili kategoriye aitse 1, değilse 0 verir.

## ÖRNEK

AraçNo	Renk
1	Kırmızı
2	Mavi
3	Kırmızı

* Dummy variable dönüşümünden sonra
AraçNo	Renk_Kırmızı	Renk_Mavi
1	        1	            0
2	        0	            1
3	        1	            0

## KRİTİK KURAL: DUMMY VARIABLE TRAP (KUKLA DEĞİŞKEN TUZAĞI)
* Dikkat edilmesi gereken en önemli şey, oluşturulan yeni sütunlardan birini dışarıda bırakmaktır.
* Eğer 2 kategori varsa (kadın/erkek), sadece biri için sütun oluşturmak yeterlidir. Çünkü bir kişi "Erkek:0" iste zaten otomatik olarak "Kadın:1" olduğu anlaşılır.
* Tüm sütunları modele eklersek, değişkenler arasında mükemmel bir bağımlılık oluşur ve bu da modelin matematiğini bozar.

# P-VALUE (OLASILIK DEĞERİ)
* Yapılan tahminin ne kadar `anlamlı` olduğunu gösteren bir ölçüttür.
* En basit tabirliyle: 'Bu sonuç tamamen şans eseri mi oldu, yoksa gerçekten bir sebebi mi var ?'

# ÖRNEK
Kahve içmek odaklanmayı arttırır mı ?
- İki ihtimal var:
1. Sıfır Hipotezi(Null hypothesis): Kahve ile odaklanma arasında hiçbir bağ yok. Bu sonuç tamamen tesadüf.
2. Alternatif Hipotezi: Kahve gerçekten odaklanmayı arttırıyor.

P-Value, sıfır hipotezinin doğru olma ihtimalidir.
**P-Value çok düşükse:** Bu sonucun tesadüf olma ihtimali sadece %1. Yani tesadüf olamaz, demek ki kahve gerçekten etkili.
**P-Value yüksekse:** Bu sonucun tesadüf olma ihtimali %50 demektir. Yani emen olamayız şans eseri olabilir.

## KRİTİK EŞİK
* Makine öğrenmesinde kabul görmüş sınır 0.05'tir (%5)
* P <0.05 ise, sonuç istatistiksel olarak anlamlıdır, tesadüf değildir.
* P> 0.05 ise, sonuç anlamlı değildir, tesadüf olabilir.


# ÖRNEK 2
Diyelim ki bir evin fiyatını tahmin ediyoruz. Regresyon modelini kurduğunda her bir değişken için ayrı bir P-Value hesaplanır.
* Metrekare (P-Value:0,001) --> Çok önemli bir değişken, modelde mutlaka kalmalı
* Binanın rengi (P-Value:0.850) --> Fiyat üzerinde hiçbir anlamlı etkisi yok.

## ÖZETLE
P-Value değeri ne kadar küçükse, o değişkene o kadar çok güveniriz.

# MAKİNE ÖĞRENMESİNDE DEĞİŞKEN SEÇİMİ

`Beş farklı yaklaşım var`

1- Bütün Değişkenleri Dahil etmek (All-in)
* Bu yöntemde seçici davranmayız, tüm verileri modelin içine atarız
* Değişken sayısı çok azsa veya hepsinin çok önemli olduğu durumlarda kullanılır


2- Geriye Doğru Eleme (Backward Elimination)
``
* Bunu bir eleme turu gibi düşünelim:
    1- Önce tüm değişkenlerle başlarız.
    2- Her birinin P-Value değerine bakarız.
    3- En yüksek P-Value değerine sahip olanı modelden çıkarırız (Anlamsız olduğunu ifade eder)
    4- Kalanlarla tekrar model kurarız
    5- Ta ki tüm değişkenlerin P-Value değeri belirlediğimiz sınırın (genelde 0.05) altına inene kadar devam ederiz.
``

3- İleriye Seçim (Forward Selection)
* Backward yönteminin tam tersidir:
    1- Sıfır değişkenle başlarız (boş küme)
    2- Değişkenleri tek tek deneriz. P-Value değeri en düşük (en anlamlı) olanı sisteme ekleriz
    3- Yeni değişkenler ekledikçe model artık iyileşmeyene kadar devam ederiz

4- İki Yönlü Eleme (Bidirectional Elimination)
* Hem ileri hem geri yöntemi aynı anda kullanır
    1- Bir değişkeni içeri alırız (İleri seçim gibi)
    2- Hemen ardından, içerideki değişkenlerden hala anlamlı mı diye kontrol ederiz, anlamsızlaşanı atarız (geriye eleme gibi)
    
5- Skor Karşılaştırması (All Possible Models / Score Comparison)
* Eldeki değişkenlerin her türlü kombinasyonu için ayrı ayrı modeller kurarız
* Modellerin başarısını ölçeriz, başarı kriterini biz belirleriz.
* En yüksek skoru veren kombinasyonu seçeriz
* En zahmetli yoldur.
