# R-SQUARED
* Bir modelin ne kadar başarılı olduğunu gösterir.
* 0 ve 1 arasındadır. Sonuç ne kadar 1'e yakınsa model o kadar başarılıdır.
* Regresyon problemlerinde kullanılır
* **from sklearn.metrics import r2_score** olarak import edilir.


![alt text](R2.png)

* Formüle göre yaptığımız işlem; modeliminiz tahmin ettiği değerin gerçek değerden farkının karesinin, eldeki verilerin ortalamasının gerçek değerden farkının karesine bölümünün 1'den çıkarılmasıdır.
* Kare almamızınm tek sebebi negatif/pozitif hatalar birbirini götürmesin ve büyük hatalar daha fazla cezalandırılsın diye.

## ÖRNEK
* Gerçek maaş değeri 10000TL
* Bizim oluşturuduğumuz model 9000TL olarak tahmin etti.
* En basit tahminin hatası ise (eldeki verilerin ortalaması) 6000TL
* Sonuç R2 = 1 - (1000² / 4000²)

# ADJUSTED R-SQUARE
* R2, modele yeni feature ekledikçe hep artıyor. Bu modelin gerçek başarısını temsil etmez.
* Bundan dolayı adjusted r-square geliştirilmiştir ve **model iyi mi ?** sorusu sormak yerine, **model gerçekten iyi mi ? yoksa şişirme mi ?** sorusunu sorar ve buna cevap arar.

* adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

![alt text](adjusted-R2.png)
* Bu formülde;
n -> Eldeki veri sayısı (kaç satır var)
p -> Feature sayısı (kaç sütun kullandık)

```python
from sklearn.metrics import r2_score
import pandas as pd
data =pd.read_csv('r2_practice_dataset.csv')
data
```

```python
# Kodda veriler direkt verilmiştir hiçbir temizleme ve makine öğrenmesi prediction yapılmamıştır.
# Örnek olması amacıyla böyle yaptım


y_true = data[['y_true']]
y_pred = data[['y_pred']]

r2 = r2_score(y_true,y_pred)
print('R2:',r2)
```

```python
# adjusted R2 hesaplama

n= len(data) #satır sayısını verir
p=1 #bağımsız değişken sayısı

adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print('R2',r2)
print('Adjusted R2',adj_r2)
```
