```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy')
#criterion='entropy' karar ağacının hangi özelliğe göre bölüneceğini seçerken kullandığı ölçüttür.
#Entropy hangi soru veriyi daha iyi ayırıyor? sorusuna cevap verir ve belirsizliği en çok azaltan split seçilir.
df = pd.read_csv('veri.csv')

#boy-kilo-yas
x = df.iloc[:,1:4].values
#cinsiyet
y = df.iloc[:,4:].values

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
dtc.fit(x_train,y_train)
y_pred = dtc.predict(x_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```
