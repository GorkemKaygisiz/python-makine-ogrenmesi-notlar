# RECEIVER OPERATING CHARACTERISTIC (ROC)
> Modelin ne kadar iyi ayırt ettiğini gösteren performans grafiğidir. True Positives ve False Positives arasındaki dengeyi görselleştirir.

## EKSENLER
**Dikey Eksen (Y)**
* True Positive Rate (Recall/Duyarlılık): Modelin gerçek '0' ları yakalama hızı. (Yukarı ne kadar yakınsa o kadar iyidir.)
**Yatay Eksen (X)**
* False Positive Rate: Modelin '1' olanlara yanlışlıkla '0' deme hızı. (Sola ne kadar yakınsa o kadar iyi)
* İdeal olan eğri sol üst köşeye en yakın olandır.

tpr(Recall) = TP / TP + FN
fpr = FP / FP + TN

# AREA UNDER CURVE (AUC)
* ROC eğrisinin altında kalan alandır. Modelin değerleri birbirinden ne kadar iyi ayırdığını ölçer.
    - AUC = 1.0: Mükemmel model
    - AUC = 0.5: Yazı-tura atar gibi tahmin eden model
    - AUC < 0.7: Zayıf model
    - AUC > 0.8: İyi model
