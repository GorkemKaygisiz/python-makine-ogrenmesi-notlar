# SUPPORT VECTOR MACHINES
* En geniş margin aralığını bularak 2 veriyi birbirinden ayıran doğruyu çizmeye çalışır.
* SVR noktaları bir tünelin içine hapsediyordu. Sınıflandırmada ayırmak için SVM bu noktaları tünel dışına iter.
* Hyperline -> Grupları ayıran ana çizgi.
* Marginal Lines -> Ana çizginin sağında ve solunda, en yakın noktalara teğet geçen çizgiler.(Sınır Çizgileri)
* Support Vectors -> Çizgiye en yakın olan, yani sınırda duran noktalardır.

# SINIFLANDIRMADA SVM KULLANIMI
* Veriler her zaman düz çizgiyle ayrılmaz, linear değildir. Burada devreye SVM kernel devreye girer.
* Linear Kernel -> Veriler bir doğruyla ayrılabiliyorsa kullanılır.
* RBF / Polynomial Kernel -> Veriler iç içe girmişse, SVM onları hayali bir 3.boyuta zıplatır.

* **from sklearn.svm import SVC** 
* **svc=SVC(kernel ='linear',random_state=0)**
* C=Classifier demek

# SVM ÖZELLİKLERİ
* KNN gibi mesafe hesapladığı için scale ister. 
* Çok büyük verilerde yavaşlayabilir ama küçük verilerde iyidir.
* c parametresi ile SVM'e bazı hataları görmezden gel, yolu geniş tut veya yolu daralt komutu verilebilir

```python
import pandas as pd
data = pd.read_csv('veri.csv')

#boy-kilo-yas
x = data.iloc[:,1:4]
#cinsiyet
y = data.iloc[:,4:]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf')
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```
