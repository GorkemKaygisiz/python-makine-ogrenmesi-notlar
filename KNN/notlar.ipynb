# KNN (K-Nearest Neighbors)
* KNN, bir verinin hangi gruba dahil olduğunu tahmin etmek için o verinin en yakın komşusuna bakar.
* Örneğin mahallede yeni taşınan birinin hangi takımı tuttuğunu bilmiyorsan, en yakın 5 komşusuna bakıp çoğunluğun tuttuğu takımı onun da tuttuğunu varsayabilriz.

## Tahmin Aşaması
- Yeni örneğe olan uzaklık tüm eğitim örneklerine hesaplanır.
- En küçük uzaklığa sahip ilk k örneği seçilir.
- Bu k komşunun çoğunluğuna göre sınıfa ata (çoğunluk oyu gibi)
- Bu sınıflandırmada k kaç komşu olacağıdır.
- K çift sayıysa mesafeye bakılabilir.

>KNN mesafe ile çalıştığı için scale yapılmalıdır.

```python
import pandas as pd
import numpy as np

data = pd.read_csv('veri.csv')

#boy-kilo-yas
x = data.iloc[:,1:4]
#cinsiyet
y = data.iloc[:,4:]

```

```python
#Modeli eğitmek için veriyi bölüyoruz
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
#KNN mesafeyle çalıştığı için veriyi ölçeklendirmemiz gerekiyor.

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
#KNN algoritmasını import ediyoruz
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5,metric ='minkowski') #default değerler bunlardır.
```

```python
#modeli eğitip tahmin yapıyoruz
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
y_pred
```

```python
#confusion matrix ile modelimizin sınıflandırma başarısını ölçüyoruz
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

![alt text](confusion-matrix.png)

* (0,0) + (1,1) doğru tahminler toplamı
* (1,0) + (0,1) yanlış tahminler toplamını verir
* Doğru ve yanlış tahminler toplamını tüm değerler toplamına bölerek modelin yüzde kaç doğru/yanlış yaptığını bulabiliriz.
