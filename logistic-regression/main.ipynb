```python
import pandas as pd 
data=pd.read_csv('veri.csv')
data.head()
```

```python
#boy-kilo-yas
x = data.iloc[:,1:4]
#cinsiyet
y = data.iloc[:,4:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(random_state=0)

logr.fit(X_train,y_train)
y_pred = logr.predict(X_test)
```

```python
print(y_test)
print(y_pred)
#Hepsini yanlış tahmin etti ama doğru çalıştı.
```

# Karmaşıklık Matrisi (Confusion Matrix)

![alt text](karmaşıklık-matrisi.png)

* True positive + True negative bize toplamda doğru bulunan değer sayısını verir. Toplam örnek sayısına bölünce yüzde kaçını doğru sınıflandırdığımızı verir.
* Aynı şekilde false positive + false negative / toplam örnek sayısı bize yanlış tahmin edilen örnekleri verir.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
#gerçek değer ve tahminler üzerinde confusion metrix oluşturduk
cm
```
