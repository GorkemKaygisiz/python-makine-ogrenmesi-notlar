# NAIVE BAYES

## BAYES TEOREMİ
* Bayes teoremi koşullu olasılıklar arasında ilişki kuran bir denklemdir. Ancak onu sıradan bir koşullu olasılık olmaktan ayıran şey **tersine çevirme gücüdür.**
* Sıradan koşullu olasılıkta genelde bir sebebin sonucunu merak ederiz, Bayes teoremi ise bize sonuçtan sebebe gitme imkanı verir.
* Uzaklık tabanlı bir algoritma olmadığı için verileri scale etmek gerekmez.
``P(sonuç/sebep) --> P(sebep/sonuç)``

$$
P(A \mid B) = \frac{P(B \mid A)\, P(A)}{P(B)}
$$
